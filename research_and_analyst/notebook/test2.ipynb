```python
import pandas as pd
import os

# === 1. File path ===
file_path = r"/home/supriyo/Downloads/1.xlsx"

# === 2. Read the Excel file ===
df = pd.read_excel(file_path, engine="openpyxl", header=None)  # No header assumed

# === 3. Extract A and E columns (0-indexed: A=0, E=4) ===
col_A = df.iloc[1:, 0].astype(str).str.strip()  # Skip A1 (header)
col_E = df.iloc[1:, 4].astype(str).str.strip()  # Skip E1 (header)

# === 4. Create sets for fast lookup ===
set_A = set(col_A.dropna())
set_E = set(col_E.dropna())

# === 5. Check for matches (A vs E) ===
df['Match_Status'] = df.iloc[:, 0].astype(str).str.strip().apply(
    lambda x: "FOUND" if x in set_E else "NOT FOUND"
)

# === 6. Materials in A but NOT in E ===
A_not_in_E = sorted(list(set_A - set_E))
df_A_not_E = pd.DataFrame({'Material_in_A_not_in_E': A_not_in_E})

# === 7. Materials in E but NOT in A ===
E_not_in_A = sorted(list(set_E - set_A))
df_E_not_A = pd.DataFrame({'Material_in_E_not_in_A': E_not_in_A})

# === 8. Combine everything and export ===
output_path = os.path.join(os.path.dirname(file_path), "match_result.xlsx")

with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Match_Status", index=False)
    df_A_not_E.to_excel(writer, sheet_name="A_not_in_E", index=False)
    df_E_not_A.to_excel(writer, sheet_name="E_not_in_A", index=False)

print(f"âœ… Matching complete.\nðŸ“„ Saved to: {output_path}")
print(f"ðŸŸ¡ {len(A_not_in_E)} items in A not found in E")
print(f"ðŸŸ  {len(E_not_in_A)} items in E not found in A")
```
